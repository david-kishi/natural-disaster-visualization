```python
import pandas as pd
import numpy as np
import datetime as df
```

```python
flood_2010s = pd.read_csv("floods_data/flood2010s.csv")
flood_2016s = pd.read_csv("floods_data/floods2016.csv")
flood_2018s = pd.read_csv("floods_data/floods2018.csv")
flood_2000s = pd.read_csv("floods_data/floods_2000s.csv")
```

```python
flood_2000s.head()
```

```python
flood_data = flood_2010s.append(flood_2016s, ignore_index = True)
```

```python
flood_data = flood_data.append(flood_2018s, ignore_index = True)
```

```python
flood_data = flood_data.append(flood_2000s, ignore_index = True)
```

```python
flood_data.head()
```

```python
flood_data.info()
```

```python
flood_data = flood_data.loc[: ,["episode_id","state","event_type","event_begin_time","event_end_time","event_timezone","damage_property","event_latitude","event_longitude","cz_type","cz_fips_code","damage_crops","source"]]
```

```python
flood_data.head()
```

```python
flood_data.info()
```

```python
flood_data[["Start_Date", "Start_Time"]] = flood_data.event_begin_time.str.split(" ", expand = True)
flood_data[["End_Date", "End_Time"]] = flood_data.event_end_time.str.split(" ", expand = True)
```

```python
flood_data
```

```python
flood_data.Start_Date = pd.to_datetime(flood_data.Start_Date)
flood_data.Start_Time = pd.to_datetime(flood_data.Start_Time)
```

```python
flood_data.End_Date = pd.to_datetime(flood_data.End_Date)
flood_data.End_Time = pd.to_datetime(flood_data.End_Time)
```

```python
new_flood = pd.DataFrame(flood_data)
```

```python
new_flood.damage_crops.fillna(0, inplace = True)
new_flood.damage_property.fillna(0, inplace = True)
```

```python
clean_floods = new_flood.to_csv(path_or_buf='floods_data.csv', index = False)
```
