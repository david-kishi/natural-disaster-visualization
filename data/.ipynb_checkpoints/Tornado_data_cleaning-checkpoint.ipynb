# 2019

```python
tor_2019=pd.read_csv("C:/Users/Daniel/Downloads/StormEvents_2019.csv")
tor_2019=tor_2019[['BEGIN_YEARMONTH','BEGIN_TIME','BEGIN_DAY','STATE',
             'EVENT_TYPE','INJURIES_DIRECT','DEATHS_DIRECT','DAMAGE_PROPERTY',
            'TOR_F_SCALE','TOR_LENGTH','TOR_WIDTH','BEGIN_LAT',
            'BEGIN_LON','END_LAT','END_LON']]

tor_2019=tor_2019[tor_2019.EVENT_TYPE == 'Tornado']
tor_2019=tor_2019.drop(columns=['EVENT_TYPE'])
tor_2019=tor_2019.fillna(0)
dates=[]

for i,r in tor_2019.iterrows():
    if len(str(r.BEGIN_DAY))==1:
        dates.append(str(r.BEGIN_YEARMONTH)[0:4]+('-')+str(r.BEGIN_YEARMONTH)[4:6]+('-0')+str(r.BEGIN_DAY))
    else:
        dates.append(str(r.BEGIN_YEARMONTH)[0:4]+('-')+str(r.BEGIN_YEARMONTH)[4:6]+('-')+str(r.BEGIN_DAY))
    

tor_2019['Date']=dates
tor_2019=tor_2019.drop(columns=['BEGIN_YEARMONTH','BEGIN_DAY'])
tor_2019.rename(columns={'STATE':'State',
                      'TOR_F_SCALE':'Fscale',
                      'INJURIES_DIRECT':'Injured',
                      'DEATHS_DIRECT':'Fatalities',
                      'DAMAGE_PROPERTY':'Property_loss',
                      'BEGIN_LAT':'SLAT',
                      'END_LAT':'ELAT',
                      'BEGIN_LON':'SLON',
                      'END_LON':'ELON',
                      'BEGIN_TIME':'Time',   
                      'TOR_LENGTH':'Length',
                      'TOR_WIDTH':'Width'},inplace=True)
tor_2019.reset_index(drop=True,inplace=True)
tor_2019['Width']=tor_2019['Width'].div(1.094)
tor_2019['Length']=tor_2019['Length'].mul(1.609)
tor_2019=tor_2019[['Date','Time','State','Fscale','Injured','Fatalities','Property_loss','SLAT','SLON','ELAT','ELON','Length','Width']]
tor_2019
```

# Function for the Others

```python
tor_dict={}
def tor_clean(year):
    df=pd.read_csv(f"C:/Users/Daniel/Downloads/StormEvents_details-ftp_v1.0_d{str(year)}.csv")
    df=df[['BEGIN_YEARMONTH','BEGIN_TIME','BEGIN_DAY','STATE',
             'EVENT_TYPE','INJURIES_DIRECT','DEATHS_DIRECT','DAMAGE_PROPERTY',
            'TOR_F_SCALE','TOR_LENGTH','TOR_WIDTH','BEGIN_LAT',
            'BEGIN_LON','END_LAT','END_LON']]
    df=df[df.EVENT_TYPE == 'Tornado']
    df=df.drop(columns=['EVENT_TYPE'])
    df=df.fillna(0)
    dates=[]

    for i,r in df.iterrows():
        if len(str(r.BEGIN_DAY))==1:
            dates.append(str(r.BEGIN_YEARMONTH)[0:4]+('-')+str(r.BEGIN_YEARMONTH)[4:6]+('-0')+str(r.BEGIN_DAY))
        else:
            dates.append(str(r.BEGIN_YEARMONTH)[0:4]+('-')+str(r.BEGIN_YEARMONTH)[4:6]+('-')+str(r.BEGIN_DAY))


    df['Date']=dates
    df=df.drop(columns=['BEGIN_YEARMONTH','BEGIN_DAY'])
    df.rename(columns={'STATE':'State',
                      'TOR_F_SCALE':'Fscale',
                      'INJURIES_DIRECT':'Injured',
                      'DEATHS_DIRECT':'Fatalities',
                      'DAMAGE_PROPERTY':'Property_loss',
                      'BEGIN_LAT':'SLAT',
                      'END_LAT':'ELAT',
                      'BEGIN_LON':'SLON',
                      'END_LON':'ELON',
                      'BEGIN_TIME':'Time',   
                      'TOR_LENGTH':'Length',
                      'TOR_WIDTH':'Width'},inplace=True)
    df.reset_index(drop=True,inplace=True)
    df['Width']=df['Width'].div(1.094)
    df['Length']=df['Length'].mul(1.609)
    df=df[['Date','Time','State','Fscale','Injured','Fatalities','Property_loss','SLAT','SLON','ELAT','ELON','Length','Width']]
    
    tor_dict[year]=df
```

```python
tor_clean(2014)
tor_clean(2015)
tor_clean(2016)
tor_clean(2017)
tor_clean(2018)
```

# Canada

```python
can_tor=pd.read_excel('C:/Users/Daniel/Downloads/Canadian_tornadoes.xlsx',sheet_name=1)
can_tor=can_tor[['YYYYMMDDHHMM_UTC','PROVINCE','FUJITA',
                 'START_LAT_N','START_LON_W','END_LAT_N','END_LON_W','LENGTH_M',
                'WIDTH_MAX_M','HUMAN_FATAL','HUMAN_INJ','DMG_THOUS',]]
can_tor=can_tor[~can_tor.END_LAT_N.isna()]
can_tor.reset_index(drop=True,inplace=True)
can_times=[]
can_dates=[]
for i in can_tor.YYYYMMDDHHMM_UTC.values:
    can_times.append(i[-4:])
    can_dates.append(i[:4]+'-'+i[4:6]+'-'+i[6:-5])
can_tor['Time']=can_times
can_tor['YYYYMMDDHHMM_UTC']=can_dates

can_tor=can_tor.fillna(0)
can_tor.rename(columns={
    'YYYYMMDDHHMM_UTC':'Date',
    'PROVINCE':'State',
    'FUJITA':'Fscale',
    'START_LAT_N':'SLAT',
    'START_LON_W':'SLON',
    'END_LAT_N':'ELAT',
    'END_LON_W':'ELON',
    'LENGTH_M':'Length',
    'WIDTH_MAX_M':'Width',
    'HUMAN_FATAL':'Fatalities',
    'HUMAN_INJ':'Injured',
    'DMG_THOUS':'Property_loss'
},inplace=True)


can_tor['Length']=can_tor['Length'].div(1000)
can_tor=can_tor[['Date','Time','State','Fscale','Injured','Fatalities','Property_loss','SLAT','SLON','ELAT','ELON','Length','Width']]
can_tor
```
